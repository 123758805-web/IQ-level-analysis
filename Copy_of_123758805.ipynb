Analysis 1: Checking the average IQ level for the UK and USA, and comparing them. this shows that the USA is on average 3 IQ points higher than the UK

```python
print("UK average:",df[df['Region'] == 'UK']['Estimated_Avg_IQ'].mean())
print("USA average:",df[df['Region'] == 'USA']['Estimated_Avg_IQ'].mean())
```

Anaysis 2: Looking at the increasing trend from the early 1900s to the present year. Here we can see that the average IQ level has been steadily increasing at a consistant rate up until the early 2000s. This is interesting as it shows that education and quality of life have a positive effect on the level of IQ globally, and that historical events, wars and illness has no effet of the increasing IQ level.
